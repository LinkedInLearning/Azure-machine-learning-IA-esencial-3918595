```python

# azureml-core of version 1.0.72 or higher is required
# azureml-dataprep[pandas] of version 1.1.34 or higher is required
from azureml.core import Workspace, Dataset
import matplotlib.pyplot as plt

subscription_id = '61decfc3-854a-4ab6-a0c8-d0c7a88bd0a9'
resource_group = 'azure-ml-ia-esencial'
workspace_name = 'workspace1'

workspace = Workspace(subscription_id, resource_group, workspace_name)

dataset = Dataset.get_by_name(workspace, name='pets')
df = dataset.to_pandas_dataframe()

avg_disease_by_species = df.groupby('species')['disease_flag'].mean()
avg_disease_by_species.plot(kind='bar', color='skyblue')
plt.title('Promedio')
plt.xlabel('Species')
plt.ylabel('Promedio')
plt.grid(axis='y')
plt.show()
```
